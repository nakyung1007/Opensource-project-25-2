# Data transformer

```python
import unicodedata, re

# Use the translations mapping you provided to rename columns
translations = {
    '거래일자': 'datetime',
    '거래시간': 'hour',
    '지역': 'region',
    '전력수요량': 'power_demand',
}

def map_col(col):
    s = str(col)
    # exact match or substring match to use the provided mapping
    for k, v in translations.items():
        if s == k or k in s:
            return v
    # fallback: normalize to ascii, remove non-alphanum and lower-case
    s2 = s.replace(' ', '_').replace('-', '_')
    s2 = unicodedata.normalize('NFKD', s2).encode('ascii', 'ignore').decode('ascii')
    s2 = re.sub(r'[^0-9A-Za-z_]', '', s2).lower()
    return s2 or col

# Build mapping from original -> new column names and apply it
new_cols = {c: map_col(c) for c in df.columns}
print('Column mapping (original -> new):')
for k, v in new_cols.items():
    print(f'{k} -> {v}')

df.rename(columns=new_cols, inplace=True)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# 그래프 스타일 설정
sns.set(style="whitegrid")

# --- 2. 데이터 읽기 ---
file_name = '/Volumes/X31/프로젝트/Opensource-project-25-2/Data/제주도_전력 수요량 시간별 데이터_2021~2023.csv'

try:
    df = pd.read_csv(file_name, encoding='utf-8')
    print(f"'{file_name}' 파일 읽기 성공 (UTF-8)")
except UnicodeDecodeError:
    df = pd.read_csv(file_name, encoding='cp949')
    print(f"'{file_name}' 파일 읽기 성공 (CP949)")
except FileNotFoundError:
    print(f"오류: '{file_name}'을(를) 찾을 수 없습니다.")
    df = None
except Exception as e:
    print(f"파일 읽기 중 오류 발생: {e}")
    df = None

if df is not None:
    # --- 3. 시간별 데이터 분석 ('거래시간' 기준) ---
    # '거래시간' 컬럼을 기준으로 그룹화
    # '전력 수요량(MWh)'의 평균(mean)을 계산합니다.
    hourly_avg = df.groupby('거래시간')['전력 수요량(MWh)'].mean().reset_index()
    
    # 컬럼 이름 변경 (시각화의 축 레이블을 위해)
    hourly_avg.rename(columns={
        '거래시간': '시간 (Hour)',
        '전력거래량(MWh)': '평균 전력거래량 (MWh)'
    }, inplace=True)
    
    print("\n--- 시간대별 평균 전력거래량 ---")
    print(hourly_avg.head()) # 데이터 확인

    # --- 4. 시각화 (막대 그래프) ---
    plt.figure(figsize=(14, 7)) # 그래프 크기 조절
    
    # seaborn의 barplot을 사용하여 시각화
    # x축: '시간 (Hour)', y축: '전력 수요량 (MWh)'
    sns.barplot(data=hourly_avg, x='시간 (Hour)', y='전력 수요량(MWh)', palette='viridis')

    # --- 5. 그래프 꾸미기 ---
    plt.title('Hourly Average Power Demand', fontsize=16, pad=20)
    plt.xlabel('Time (1am ~ 24pm)', fontsize=12)
    plt.ylabel('Mean energy(MWh)', fontsize=12)
    
    # x축 눈금 설정 (1시부터 24시까지)
    # 0부터 23까지의 인덱스를 1부터 24까지의 레이블로 변경
    plt.xticks(ticks=range(0, 24), labels=range(1, 25))
    
    # 그래프 표시
    plt.show()
```
