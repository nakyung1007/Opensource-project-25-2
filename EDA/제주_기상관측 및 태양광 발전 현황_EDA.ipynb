## 데이터 불러오기

```python
import pandas as pd
```

```python

raw_path = "/Volumes/X31/프로젝트/에너지 AI/Opensource-project-25-2/Data/real/한국동서발전(주)_제주 기상관측 및 태양광 발전 현황_20240531.csv"
try:
    df = pd.read_csv(raw_path, encoding="utf-8")
except UnicodeDecodeError:
    try:
        df = pd.read_csv(raw_path, encoding="cp949")
    except Exception:
        df = pd.read_csv(raw_path, encoding="euc-kr", errors="replace")

print(df.info())
print(df.head())
```

```python
# 날짜 컬럼을 datetime으로 변환 (첫 번째 컬럼이 날짜라고 가정)
date_column = df.columns[0]  # 첫 번째 컬럼명 자동 감지
df[date_column] = pd.to_datetime(df[date_column])

# 2024년 데이터만 필터링
df_2024 = df[df[date_column].dt.year == 2024]

# 필터링된 결과 확인
print(f"전체 데이터 개수: {len(df)}")
print(f"2024년 데이터 개수: {len(df_2024)}")
print("\n2024년 데이터 샘플:")
print(df_2024.head())

# 새로운 CSV 파일로 저장
df_2024.to_csv('2024년_기상데이터.csv', index=False, encoding='utf-8-sig')
print("\n'2024년_기상데이터.csv' 파일로 저장 완료!")
```
